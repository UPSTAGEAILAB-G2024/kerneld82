```python
import pandas as pd

df = pd.read_excel('116_DT_MLTM_5373_20240502120559.xls.xlsx')
df
```

```python
colLength = len(df.columns)
colLength
```

```python
for y in range(100):
    for m in range(12):
        col = y * 12 + m
        
        if col >= colLength:
            break
        
        total = df.iloc[0, col]
        apt = df.iloc[5, col]
        notApt = total - apt
        
        print(f"{y + 2018}\t{m + 1}\t{apt}\t{notApt}\t{total}")
        
```
