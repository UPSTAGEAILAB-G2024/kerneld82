```python
import pandas as pd
df = pd.read_csv('../data/서울시_가계대출규모/원본/가계대출규모_20240506084513.csv')
df.head()
```

```python
# 합계 컬럼 추가
df['합계'] = df['예금취급기관'] + df['예금은행'] + df['비은행예금취급기관']
df.head()
```

```python
ymList = list(map(lambda x: df['시점'][x].split('. '), range(len(df))))
ymList
```

```python
yList = []

for each in ymList:
    yList.append(int(each[0]))

yList
```

```python
mList = []

for each in ymList:
    mList.append(int(each[1]))

mList
```

```python
df['기준년도'] = yList
df['기준월'] = mList
df.head()
```

```python
df = df.drop('시점', axis=1)
df.head()
```

```python
df['지역'] = '서울'
df.head()
```

```python
df.to_csv('../data/서울시_가계대출규모/전처리/전처리_서울시_가계대출규모.csv', encoding='utf-8-sig')
```
