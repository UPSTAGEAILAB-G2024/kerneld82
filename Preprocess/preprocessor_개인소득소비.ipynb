```python
import pandas as pd
df = pd.read_csv('../data/서울시_1인당_지역내총생산_지역총소득_개인소득/원본/copy.csv')
df
```

```python
colLength = len(df.columns)
colLength
```

```python
yearList = []
localProduceList = []
localEarnList = []
personalEarnList = []
personalConsumeList = []

for i in range(10000000):
    col1 = 4 * i + 1
    col2 = 4 * i + 2
    col3 = 4 * i + 3
    col4 = 4 * i + 4

    if col1 >= colLength or \
       col2 >= colLength or \
       col3 >= colLength or \
       col4 >= colLength:
        break
    
    yearList.append(int(df.columns[col1]))
    localProduceList.append(int(df.iloc[1, col1]))
    localEarnList.append(int(df.iloc[1, col2]))
    personalEarnList.append(int(df.iloc[1, col3]))
    personalConsumeList.append(int(df.iloc[1, col4]))
```

```python
df2 = pd.DataFrame()
df2['기준년도'] = yearList
df2['1인당 지역내총생산'] = localProduceList
df2['1인당 지역총소득'] = localEarnList
df2['1인당 개인소득'] = personalEarnList
df2['1인당 민간소비'] = personalConsumeList
df2
```

```python
df2['가상? 가처분소득'] = df2['1인당 개인소득'] - df2['1인당 민간소비']
df2
```

```python
df2.to_csv('../data/서울시_1인당_지역내총생산_지역총소득_개인소득/전처리/전처리_개인소득소비.csv', encoding = 'utf-8-sig')
```
