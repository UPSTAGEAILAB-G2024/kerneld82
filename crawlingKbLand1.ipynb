# KB부동산에서 전국매수우위지수를 크롤링

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

ChromeDriverManager().install()
browser = webdriver.Chrome()
```

```python
url = 'https://data.kbland.kr/kbstats/psychology-of-housing-market'
browser.get(url)
```

```python
from selenium.webdriver.common.by import By
```

```python
# '10년' 라디오 버튼 클릭
radioList = browser.find_elements(By.CLASS_NAME, 'custom-radio')
radioList[4].click()
```

```python
elem = browser.find_element(By.CLASS_NAME, 'tradingtable')
tableRows = elem.find_elements(By.CLASS_NAME, 'tablelow')
```

```python
# 년월 목록 가져오기
spanYmElemList = tableRows[0].find_elements(By.TAG_NAME, 'span')

ymList = []
for each in spanYmElemList:
    ymList.append(each.text)
    print(each.text)
```

```python
# 매수우위지수 가져오기
buySuperiorityIdxElemList = tableRows[4].find_elements(By.TAG_NAME, 'span')

buySuperiorityIdxList = []
for each in buySuperiorityIdxElemList:
    buySuperiorityIdxList.append(each.text)
    print(each.text)
```

## 크롤링한 데이터를 csv 파일로 만들기


```python
yearList = []
monthList = []

for each in ymList:
    each = each.replace("'", "")
    tmp = each.split('.')
    
    year = int("20" + tmp[0])
    month = int(tmp[1])
    
    yearList.append(year)
    monthList.append(month)
    
print(yearList)
print(monthList)

print(len(yearList))
print(len(monthList))
```

```python

import pandas as pd

df = pd.DataFrame(data = {'기준년도' : yearList, '기준월' : monthList, '매수우위지수' : buySuperiorityIdxList})
df
```

```python
# 2018년 이상 데이터만 살리기
mask = df['기준년도'] >= 2018
df = df[mask]
df
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_csv('data/매수우위지수.csv', encoding='utf-8-sig')
```
