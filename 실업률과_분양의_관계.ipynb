```python
import pandas as pd

# 실업률 테이블 로드
df1 = pd.read_csv('data/temp/실업률.csv')
df1
```

```python
# 연도-월 컬럼 추가
ymList = list(map(lambda x: '%d-%02d' % (int(df1['기준년도'][x]), int(df1['기준월'][x])), range(len(df1))))
df1['ym'] = pd.DataFrame(ymList)
df1
```

```python
# 실업률 컬럼 추가
df1['실업률'] = df1['경제활동인구_실업자_합계'] / df1['경제활동인구_합계'] * 100
df1
```

```python
# 분양실적
df2 = pd.read_csv('data/공급량_관련/전처리/전처리_주택건설_분양실적_공동주택(월별_순계)_호.csv')
df2
```

```python
# 연도-월 컬럼 추가
ymList = list(map(lambda x: '%d-%02d' % (int(df2['기준년도'][x]), int(df2['기준월'][x])), range(len(df2))))
df2['ym'] = pd.DataFrame(ymList)
df2
```

```python
# 분양 합계 값을 실업률과 동일한 스케일로 변경.
# 실업률과 분양 합계를 하나의 그래프에서 같이 보여주고 싶어서.
maxUnemploymentRate = df1['실업률'].max()
maxSum = df2['합계'].max()

df2['scaledSellSum'] = df2['합계'] / maxSum * maxUnemploymentRate
df2
```

```python
df3 = pd.merge(left = df1, right = df2, on = ['ym'], how = 'inner')
df3
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 폰트 목록
import matplotlib.font_manager
print([f.name for f in matplotlib.font_manager.fontManager.ttflist])
```

```python
plt.figure(figsize=(15, 4))
sns.lineplot(data = df3, x = 'ym', y = '실업률', label='실업률')
sns.lineplot(data = df3, x = 'ym', y = 'scaledSellSum', label='분양합계')
plt.xlabel('년월')
plt.xticks(rotation = 90)
plt.ylabel('')
plt.show()
```
